4. Count the number of duplicates in the Google Play dataset. You must build a function to loop through the Google Play dataset

````python
# Solution

def contar_duplicados(google):

    total_filas = len(google)

    filas_unicas = len(google.drop_duplicates())
    filas_duplicadas = total_filas - filas_unicas
    return filas_duplicadas

num_duplicados = contar_duplicados(google)
print("N√∫mero de filas duplicadas: ", num_duplicados)

````

6. Remove non-english apps from both Android and Apple datasets. You should write a function that takes a string, if there's more than three characters that fall outside the ASCII range (0 - 127), then the function should return False (identify the string as non-English), otherwise it should return True. Hint: use the [```ord()```  function](https://www.w3schools.com/python/ref_func_ord.asp)

````python
# For instance
correct_english_name = "Instachat üòú"  # 1 character outside the ASCII range (üòú)
incorrect_english_name = "Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠" # 13 character outside the ASCII range (ie Êí≠)
````

````python
print(ord("I"))
print(ord("t"))
print(ord("-"))
print(ord("üòú"))
print(ord("Áà±"))
````

````python
# Solution
def es_ingles(google):
    caracteres_no_ascii = 0

    for caracter in google:
        if ord(caracter) > 127:
            caracteres_no_ascii += 1

    if caracteres_no_ascii > 3:
        return False
    else:
        return True
````

````python
android_ingles = google[google['App'].apply(es_ingles)]
apple_ingles = apple[apple['track_name'].apply(es_ingles)]
android_ingles
````

 8. Build a frequency table for the ```prime_genre``` column of the App Store dataset, and the ```Genres``` and ```Category``` columns of the Google Play data set. What are the most common genres for each market? You must develop and deploy your function

````python
def tabla_frecuencias(df, columna):

    frecuencias = {}

    for index, row in df.iterrows():
        valor = row[columna]


        if valor in frecuencias:
            frecuencias[valor] += 1
        else:
            frecuencias[valor] = 1

    return frecuencias

frecuencias_apple = tabla_frecuencias(apple, 'prime_genre')
frecuencias_google_genres = tabla_frecuencias(google, 'Genres')
frecuencias_google_category = tabla_frecuencias(google, 'Category')

frecuencias_apple
````

````python
frecuencias_google_category
````

````python
genero_mas_comun_apple = max(frecuencias_apple, key=frecuencias_apple.get)
genero_mas_comun_google_genres = max(frecuencias_google_genres, key=frecuencias_google_genres.get)
genero_mas_comun_google_category = max(frecuencias_google_category, key=frecuencias_google_category.get)

print("G√©nero m√°s com√∫n en la App Store: ", genero_mas_comun_apple)
print("G√©nero m√°s com√∫n en Google Play (Genres): ", genero_mas_comun_google_genres)
print("G√©nero m√°s com√∫n en Google Play (Category): ", genero_mas_comun_google_category)
````

 9. One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. Based on that exercise, give one app profile recommendation for Google Play.

    For the Google Play data set, we can find the required information in the ```Installs``` column. Then, compute the average number of installs by genre. Hint 1: consider only unique values of the ```Category``` column. Hint 2: If an app has, for instance, ```1,000,000+``` assume it as ```1,000,000``` installs. Hint 3: Be careful with a skewed distribution

````python
# Solution
instalaciones_por_genero = {}

for index, row in google.iterrows():
    genero = row['Category']
    instalaciones = row['Installs']

    instalaciones = instalaciones.replace('+', '')
    instalaciones = instalaciones.replace(',', '')
    instalaciones = float(instalaciones)

    if genero in instalaciones_por_genero:
        instalaciones_por_genero[genero][0] += instalaciones
        instalaciones_por_genero[genero][1] += 1
    else:
        instalaciones_por_genero[genero] = [instalaciones, 1]

promedio_instalaciones_por_genero = {genero: total_instalaciones / num_apps for genero, (total_instalaciones, num_apps) in instalaciones_por_genero.items()}
promedio_instalaciones_por_genero
````

````python
genero_mas_popular = max(promedio_instalaciones_por_genero, key=promedio_instalaciones_por_genero.get)

print("G√©nero m√°s popular en Google Play: ", genero_mas_popular)
````

 10. The App Store dataset doesn¬¥t have the ```installs``` information. Then, you will construct the average rating from the ```rating_count_tot``` column per genre. Based on that exercise, give one app profile recommendation for the App Store

````python

calificaciones_por_genero = {}


for index, row in apple.iterrows():
    genero = row['prime_genre']
    calificaciones = row['rating_count_tot']


    if genero in calificaciones_por_genero:
        calificaciones_por_genero[genero][0] += calificaciones
        calificaciones_por_genero[genero][1] += 1

    else:
        calificaciones_por_genero[genero] = [calificaciones, 1]

promedio_calificaciones_por_genero = {genero: total_calificaciones / num_apps for genero, (total_calificaciones, num_apps) in calificaciones_por_genero.items()}

genero_mas_popular = max(promedio_calificaciones_por_genero, key=promedio_calificaciones_por_genero.get)

print("G√©nero m√°s popular en la App Store: ", genero_mas_popular)
````
